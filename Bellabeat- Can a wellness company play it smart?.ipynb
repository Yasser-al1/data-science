### Graphs 1-10:

**Graph 1: Regardless of distance walked, more bed time means more sleep.**

```r
# plot data and  widen it
options(repr.plot.width=15, repr.plot.height=8)
ggplot(acivity_and_sleep, aes(x=TotalTimeInBed,
                              y=TotalMinutesAsleep, 
                              color=TotalDistance))+
    geom_point(size=3)+
    geom_smooth(method = "lm", formula= y~x, colour = "black")
```

**Graph 2: More steps means slightly less sleep**

More steps may indicate business (or sleeping less due to work/home circumstances).

```r
# plot data and  widen it
options(repr.plot.width=15, repr.plot.height=8)
ggplot(acivity_and_sleep, aes(x=TotalSteps,
                              y=TotalMinutesAsleep, 
                              color=TotalTimeInBed))+
    geom_point(size=3)+
    geom_smooth(method = "lm", formula= y~x, colour = "black")
```

**Graph 3: Running & sleep don't seem to have a relationship**
